```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
url  = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris_data = pd.read_csv(url, names=column_names)
```

```python
iris_data.head()
```

```python
iris_data.describe()
```

```python
sns.pairplot(iris_data, hue="class")
plt.show()
```

```python
x = iris_data.drop("class", axis = 1)
x
```

```python
y = iris_data["class"]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```

```python
# prompt: knn classifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# prompt: performing feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Print the scaled data
print(x_scaled)
```

```python
# prompt: performing dimensionality reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_reduced = pca.fit_transform(x_scaled)

# Print the reduced data
print(x_reduced)
```

```python
# prompt: visualizing the data in 2D

import matplotlib.pyplot as plt
# Map the categorical labels to numerical values for plotting
# This assumes 'y' contains the labels
y_colors = {
    'Iris-setosa': 'red',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'blue'
}
y_mapped = [y_colors[label] for label in y]

plt.scatter(x_reduced[:, 0], x_reduced[:, 1], c=y_mapped)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
